```python
from smartcapex_dv import schema_tools as tools
oss3g
```

```python
oss3g_schema = tools.infer_schema(data = oss3g)
oss3g_schema = tools.set_batch_id_column(oss3g_schema, 'week_period')
oss3g_schema = tools.set_max_fraction_threshold(oss3g_schema, 1.3)
oss3g_schema = tools.set_min_fraction_threshold(oss3g_schema, .7)
oss3g_schema = tools.set_unicity_features(oss3g_schema, 'WCEL name')
oss3g_schema = tools.set_regex(oss3g_schema, 'WCEL name', '^(SMALL|OCI\d{4})_.*_U\d{1,2}-\d\d?$')
oss3g_schema = tools.set_type(oss3g_schema, 'week_period', 'int')
oss3g_schema = tools.set_domain(oss3g_schema, 'week_period', '')
oss3g_schema = tools.set_regex(oss3g_schema, 'week_period', '^202\d([0-4]\d|5[0-3])$')
positive_value_columns = ['WBTS ID', 'busyhour_DOWNLOADED_TRAFFIC',
                           'sum_DOWNLOADED_TRAFFIC', 'busyhour_UPLOADED_TRAFFIC',
                           'sum_UPLOADED_TRAFFIC', 'busyhour_DOWNLOADED_THROUGHPUT',
                           'avg_DOWNLOADED_THROUGHPUT', 'busyhour_average_number_of_users_dl',
                           'average_number_of_users_dl', 'busyhour_average_number_of_users_ul',
                           'average_number_of_users_ul', 'busyhour_DL_Power_Load',
                           'average_DL_Power_Load', 'busyhour_average_number_of_users_in_queue',
                           'average_number_of_users_in_queue']
for i,column in enumerate(oss3g.columns):
    if tools.get_feature_presence(oss3g_schema, column) != 1: 
        oss3g_schema = tools.set_presence(oss3g_schema, column, .7)
    if i > 0:
        oss3g_schema = tools.set_type(oss3g_schema, column, 'float')
    if i > 2:
        oss3g_schema = tools.set_outliers(oss3g_schema, column, 7)
    if column in positive_value_columns:
        oss3g_schema = tools.set_domain (oss3g_schema, column, min_value=0)
        
oss3g_schema = tools.set_mapped_to_only_one(oss3g_schema, 'WBTS ID', 'WCEL name')
oss3g_schema = tools.add_custom_constraint(oss3g_schema, 'cell_name_column', 'WCEL name')
oss3g_schema = tools.add_custom_constraint(oss3g_schema, 'max_new_cells', 300)
oss3g_schema = tools.add_custom_constraint(oss3g_schema, 'max_disappeared_cells', 200)
```

```python
tools.get_compact_schema(oss3g_schema)
```

```python
oss4g_schema = tools.infer_schema(data = oss4g)
oss4g_schema = tools.set_batch_id_column(oss4g_schema, 'week_period')
oss4g_schema = tools.set_max_fraction_threshold(oss4g_schema, 1.3)
oss4g_schema = tools.set_min_fraction_threshold(oss4g_schema, .7)
oss4g_schema = tools.set_unicity_features(oss4g_schema, 'CELL')
oss4g_schema = tools.set_regex(oss4g_schema, 'CELL', '^(SMALL|OCI\d{4})_.*_L\d{1,2}-\d\d?$')
oss4g_schema = tools.set_type(oss4g_schema, 'week_period', 'int')
oss4g_schema = tools.set_domain(oss4g_schema, 'week_period', '')
oss4g_schema = tools.set_regex(oss4g_schema, 'week_period', '^202\d([0-4]\d|5[0-3])$')

for i,column in enumerate(oss4g.columns):
    if tools.get_feature_presence(oss4g_schema, column) != 1: 
        oss4g_schema = tools.set_presence(oss4g_schema, column, .7)
    if i > 0:
        oss4g_schema = tools.set_type(oss4g_schema, column, 'float')
        if 'percentage' in column:
            oss4g_schema = tools.set_domain (oss4g_schema, column, min_value=0, max_value=100)
        else : 
            oss4g_schema = tools.set_domain (oss4g_schema, column, min_value=0)
    if i > 1:
        oss4g_schema = tools.set_outliers(oss4g_schema, column, 7)
        
oss4g_schema = tools.add_custom_constraint(oss4g_schema, 'cell_name_column', 'CELL')
oss4g_schema = tools.add_custom_constraint(oss4g_schema, 'max_new_cells', 300)
oss4g_schema = tools.add_custom_constraint(oss4g_schema, 'max_disappeared_cells', 200)
```
