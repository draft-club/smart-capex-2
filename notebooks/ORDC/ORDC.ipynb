```python
import pandas as pd
import src.schema_tools as tools
from decimal import Decimal
pd. set_option('display.max_columns', None)
```

```python
oss2G = pd.read_csv('/mnt/data/ORDC/02_intermediate/preprocessed_oss_counter_2g.csv', delimiter='|')
oss3G = pd.read_csv('/mnt/data/ORDC/02_intermediate/preprocessed_oss_counter_3g.csv', delimiter='|')
oss4G = pd.read_csv('/mnt/data/ORDC/02_intermediate/preprocessed_oss_counter_4g.csv', delimiter='|')
```

```python
schema_oss2G = tools.infer_schema(data=oss2G)
schema_oss3G = tools.infer_schema(data=oss3G)
schema_oss4G = tools.infer_schema(data=oss4G)

###################### 2G ######################
tools.set_unicity_features(schema_oss2G, ['week_period','cell_id'])
tools.set_max_fraction_threshold(schema_oss2G, 1.1)
tools.set_min_fraction_threshold(schema_oss2G, 0.9)

tools.set_type(schema_oss2G, feature_name='week_period', feature_type= 'str')
tools.set_regex(schema_oss2G, feature_name='week_period', regex= '^#year#0?#week#')

tools.set_regex(schema_oss2G, feature_name='cell_id', regex= '^[0-9]*$')
tools.set_mapped_to_only_one(schema_oss2G, feature_name='cell_id', mapped_to_only_one='cell_name')

tools.set_regex(schema_oss2G, feature_name='cell_name', regex= '^[A-Z]{2}_[0-9]{2}_[0-9]{3,4}_.{2}_.{2,4}$')
tools.set_mapped_to_only_one(schema_oss2G, feature_name='cell_name', mapped_to_only_one='cell_id')

tools.set_type(schema_oss2G, feature_name= 'year', feature_type= 'str')
tools.set_regex(schema_oss2G, feature_name = 'year', regex= '^202[0-9]$')

tools.set_type(schema_oss2G, feature_name= 'week', feature_type= 'str')
tools.set_regex(schema_oss2G, feature_name = 'week', regex= '^[1-5]?[0-9]$')
     
tools.add_custom_constraint(schema_oss2G, "max_new_cells", 300)
tools.add_custom_constraint(schema_oss2G, "max_disappeared_cells", 200)
tools.add_custom_constraint(schema_oss2G, "max_cell_per_site", 24)

###################### 3G ######################
tools.set_unicity_features(schema_oss3G, ['week_period','cell_id'])
tools.set_max_fraction_threshold(schema_oss3G, 1.1)
tools.set_min_fraction_threshold(schema_oss3G, 0.9)

tools.set_type(schema_oss3G, feature_name='week_period', feature_type= 'str')
tools.set_regex(schema_oss3G, feature_name='week_period', regex= '^#year#0?#week#')

tools.set_regex(schema_oss3G, feature_name='cell_id', regex= '^[0-9]*$')
tools.set_mapped_to_only_one(schema_oss3G, feature_name='cell_id', mapped_to_only_one='cell_name')

tools.set_regex(schema_oss3G, feature_name='cell_name', regex= '^[A-Z]{2}_[0-9]{2}_[0-9]{3,4}_.{2}_.{2,4}$')
tools.set_mapped_to_only_one(schema_oss3G, feature_name='cell_name', mapped_to_only_one='cell_id')

tools.set_type(schema_oss3G, feature_name= 'year', feature_type= 'str')
tools.set_regex(schema_oss3G, feature_name = 'year', regex= '^202[0-9]$')

tools.set_type(schema_oss3G, feature_name= 'week', feature_type= 'str')
tools.set_regex(schema_oss3G, feature_name = 'week', regex= '^[1-5]?[0-9]$')
     
tools.add_custom_constraint(schema_oss3G, "max_new_cells", 300)
tools.add_custom_constraint(schema_oss3G, "max_disappeared_cells", 200)
tools.add_custom_constraint(schema_oss3G, "max_cell_per_site", 24)

###################### 4G ######################
tools.set_unicity_features(schema_oss4G, ['week_period','cell_id'])
tools.set_max_fraction_threshold(schema_oss4G, 1.1)
tools.set_min_fraction_threshold(schema_oss4G, 0.9)

tools.set_type(schema_oss4G, feature_name='week_period', feature_type= 'str')
tools.set_regex(schema_oss4G, feature_name='week_period', regex= '^#year#0?#week#$')

tools.set_regex(schema_oss4G, feature_name='cell_id', regex= '^[0-9]*$')
tools.set_mapped_to_only_one(schema_oss4G, feature_name='cell_id', mapped_to_only_one='cell_name')

tools.set_regex(schema_oss4G, feature_name='cell_name', regex= '^[A-Z]{2}_[0-9]{2}_[0-9]{3,4}_.{2}_.{2,4}$')
tools.set_mapped_to_only_one(schema_oss4G, feature_name='cell_name', mapped_to_only_one='cell_id')

tools.set_type(schema_oss4G, feature_name= 'year', feature_type= 'str')
tools.set_regex(schema_oss4G, feature_name = 'year', regex= '^202[0-9]$')

tools.set_type(schema_oss4G, feature_name= 'week', feature_type= 'str')
tools.set_regex(schema_oss4G, feature_name = 'week', regex= '^[1-5]?[0-9]$')
     
tools.add_custom_constraint(schema_oss4G, "max_new_cells", 300)
tools.add_custom_constraint(schema_oss4G, "max_disappeared_cells", 200)
tools.add_custom_constraint(schema_oss4G, "max_cell_per_site", 24)
```

```python
tools.save_schema(schema_oss2G, 'oss2G_schema.json')
tools.save_schema(schema_oss3G, 'oss3G_schema.json')
tools.save_schema(schema_oss4G, 'oss4G_schema.json')
```

```python
sites_list = pd.read_excel('/mnt/data/ORDC/01_raw/deployment_history/Liste existing site ORDC 31032022.xlsx')
sites_list
```

```python
sites_list_schema = tools.infer_schema(data=sites_list, max_domain_values= 50)
tools.set_unicity_features(sites_list_schema, 'Key')

tools.set_regex(sites_list_schema, feature_name= 'Key', regex= '^.*_(Orange|Tigo)$')

tools.set_type(sites_list_schema, feature_name= 'Site ID', feature_type= "str")
tools.set_distinctness(sites_list_schema, feature_name= 'Site ID', distinctness_condition='eq', distinctness_value= 1)
tools.set_mapped_to_only_one(sites_list_schema, feature_name= 'Site ID', mapped_to_only_one= ['Site Name', 'Site Code'])
tools.set_regex(sites_list_schema, feature_name= 'Site ID', regex='^[0-9]*$')

tools.set_type(sites_list_schema, feature_name= 'Site Name', feature_type= "str")
tools.set_distinctness(sites_list_schema, feature_name= 'Site Name', distinctness_condition='eq', distinctness_value= 1)
tools.set_mapped_to_only_one(sites_list_schema, feature_name= 'Site Name', mapped_to_only_one= ['Site ID', 'Site Code'])

tools.set_type(sites_list_schema, feature_name= 'Site Code', feature_type= "str")
tools.set_distinctness(sites_list_schema, feature_name= 'Site Code', distinctness_condition='eq', distinctness_value= 1)
tools.set_regex(schema_oss4G, feature_name='Site Code', regex= '^[A-Z]{2}_[0-9]{2}_[0-9]{3,4}_.{2}$')
tools.set_mapped_to_only_one(sites_list_schema, feature_name= 'Site Code', mapped_to_only_one= ['Site Name', 'Site Code'])


tools.set_domain(sites_list_schema, feature_name='Longitude', min_value=11, max_value= 32)

tools.set_type(sites_list_schema, feature_name='Latitude', feature_type= 'float')
tools.set_domain(sites_list_schema, feature_name='Latitude', min_value=-14, max_value= 5)

tools.set_presence(sites_list_schema, feature_name='Coverage\ntype', presence = 1)
tools.set_domain(sites_list_schema, feature_name='Coverage\ntype', 
                 domain= [x for x in sites_list['Coverage\ntype'].unique() if x == x] )

tools.set_regex(sites_list_schema, feature_name= 'Year', regex='^202[0-5]$')
tools.set_type(sites_list_schema, feature_name= 'Year', feature_type="str")

pass

```

```python
tools.save_schema(sites_list_schema, 'sites_list_schema.json')
```

```python
cells_list = pd.read_excel('/mnt/data/ORDC/01_raw/deployment_history/CELLES MACROS SITE OCD.xlsx')
cells_list
```

```python
cells_list_schema = tools.infer_schema(data=cells_list, max_domain_values= 50)

tools.set_regex(cells_list_schema, feature_name= 'Key', regex= '^.*_(Orange|Tigo)$')

tools.set_regex(cells_list_schema, feature_name='Site ID\nNodeB ID', regex = '^[1-9][0-9]{3,4}$')

tools.set_regex(cells_list_schema, feature_name='Site Code\nNodeB Code', regex= '^[A-Z]{2}_[0-9]{2}_[0-9]{3,4}_.{2}$')

tools.set_regex(cells_list_schema, feature_name='Cell Code', regex= '^[A-Z]{2}_[0-9]{2}_[0-9]{3,4}_.{2}_.{2,4}$')

##### not always true (upped_lower, _1234 ...)
tools.set_regex(cells_list_schema, feature_name='Cell Name', regex= '^#Site Name \nNodeB Name#_.{2,4}$')

tools.set_type(cells_list_schema, feature_name='MCC', feature_type= 'str')
tools.set_domain(cells_list_schema, feature_name='MCC', domain= ['630'] )


tools.set_type(cells_list_schema, feature_name='MNC', feature_type= 'str')
tools.set_domain(cells_list_schema, feature_name='MNC', domain= ['86','89'] )


tools.set_type(cells_list_schema, feature_name='LAC', feature_type= 'str')
tools.set_regex(cells_list_schema, feature_name='LAC', regex = '^[1-9][0-9]{4}$' )

tools.set_type(cells_list_schema, feature_name='Cell ID\nECI (4G)', feature_type= 'str')
tools.set_regex(cells_list_schema, feature_name='Cell ID\nECI (4G)', regex = '^[0-9]{3,5}$' )

tools.set_type(cells_list_schema, feature_name='CGI \nECGI (4G)', feature_type= 'str')
tools.set_regex(cells_list_schema, feature_name='CGI \nECGI (4G)', regex = '#MCC##MNC##LAC#0{0,3}#Cell ID\nECI (4G)#$' )

tools.set_type(cells_list_schema, feature_name='CGI (Otarie)', feature_type= 'str')
tools.set_regex(cells_list_schema, feature_name='CGI (Otarie)', regex = '#MCC#-#MNC#-#LAC#-#Cell ID\nECI (4G)#$' )

tools.set_domain(cells_list_schema, feature_name='Longitude', min_value=11, max_value= 32)

tools.set_type(cells_list_schema, feature_name='Latitude', feature_type= 'float')
tools.set_domain(cells_list_schema, feature_name='Latitude', min_value=-14, max_value= 5)

tools.set_mapped_to_only_one(cells_list_schema, feature_name= 'BSC Name\nRNC Name', mapped_to_only_one= 'BSC Code\nRNC Code')
tools.set_mapped_to_only_one(cells_list_schema, feature_name= 'BSC Code\nRNC Code', mapped_to_only_one= 'BSC Name\nRNC Name')

pass
```
