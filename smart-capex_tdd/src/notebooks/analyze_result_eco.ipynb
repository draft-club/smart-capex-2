```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/OMA/TDD/05_models_output/final_npv/20240529_115527/final_npv_of_the_upgrade_FDD_from_capacity.csv',sep='|')
df = pd.read_csv('data/OMA/TDD/05_models_output/final_npv/20240531_153936/final_npv_of_the_upgrade_TDD_from_capacity.csv',sep='|')
df
```

```python
npv = df[['site_id', 'NPV']]
npv
```

```python
npv_sort = npv.sort_values(by='NPV', ascending=False)
npv_sort
```

```python
npv_sort['cumsum'] = npv_sort['NPV'].cumsum()
npv_sort = npv_sort.reset_index()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(npv_sort['cumsum'], marker='o')
plt.title('Graphique de la Somme Cumulée')
plt.xlabel('Index')
plt.ylabel('Somme Cumulée')
plt.grid(True)
plt.show()
```

```python
positif = (npv_sort['NPV'] > 0).sum()
negatif = (npv_sort['NPV'] < 0).sum()

# Prepare Data
counts = {'Positif NPV': positif, 'Negatif NPV': negatif}
labels = list(counts.keys())
values = list(counts.values())

# Figure
plt.figure(figsize=(10, 6))
plt.bar(labels, values, color=['green', 'red'])
plt.title('Positif and Negatif NPV')
plt.xlabel('Val')
plt.ylabel('Number')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

bars = ax.bar(labels, values, color=['green', 'red'], edgecolor='black')

# Ajouter les valeurs sur les barres
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', fontsize=12, fontweight='bold')

# Personnaliser le graphique
ax.set_title('Nombre de Valeurs Positives et Négatives', fontsize=16, fontweight='bold')
ax.set_xlabel('Type de Valeur', fontsize=14)
ax.set_ylabel('Nombre', fontsize=14)
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_axisbelow(True)

# Ajouter des bordures arrondies aux barres
for bar in bars:
    bar.set_linewidth(1.5)
    bar.set_edgecolor('black')
    bar.set_linestyle('-')

plt.tight_layout()
plt.show()
```
